```python
import pandas as pd
from sportsref.baseball import Season
import numpy as np
#ploting libs
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
szn2021 = Season(2021)
```

```python
szn2021.overview()
# you can grab the table from the link below instead of using my sportsref library
```

```python
df = szn2021.overview().get_df("team_output")
```

```python
cols_to_keep  = ["SP", "RP", "C", "1B", "2B", "3B", "SS","LF","CF","RF","DH"]
```

```python
df
```

```python
pos_df = df[cols_to_keep]
```

```python
def split_col(col):
    return [col.str.slice(0,3), col.str.slice(3)]
```

```python
def melt_war_table(df,cols):
    dfs = []
    for pos in cols:
        col_team, col_val = split_col(df[pos])
        temp_df = pd.DataFrame({'team' : col_team, 'war': col_val})
        temp_df['position'] = pos
        temp_df['war'] = pd.to_numeric(temp_df['war'])
        dfs.append(temp_df)
    return pd.concat(dfs, ignore_index=True)
```

```python
melted_df = melt_war_table(df,cols_to_keep)
melted_df
```

```python
plot_df = melted_df.pivot(index='team', columns='position', values='war').reindex(columns=cols_to_keep)
plot_df
```

```python
# let's plot
fig, ax = plt.subplots(figsize=(16,15))
im = ax.imshow(plot_df, cmap="bwr",
               vmin=-3,vmax=3,
              aspect='auto')

#colorbar
# cmap = mpl.cm.bwr
# norm = mpl.colors.Normalize(vmin=-3,vmax=3)
cbar = ax.figure.colorbar(im,
                          ax=ax, extend='both')
#axes
ax.set_xticks(np.arange(len(plot_df.columns)))
ax.set_yticks(np.arange(len(plot_df.index)))
ax.tick_params(
    top=True,
    labeltop=True)
# labels
ax.set_xticklabels(plot_df.columns, size=20)
ax.set_yticklabels(plot_df.index, size=20)

#loop over data dimentions and create text annotations
for i in range(len(plot_df.columns)):
    for j in range(len(plot_df.index)):
        text = ax.text(i,j,'{0:2g}'.format(plot_df.iloc[j,i]),
                      ha="center", va="center", color="k", size=15)
        

# ax.set_aspect('auto')        
ax.set_title("WAR(b) from position", size=30, pad=20)
# sign
plt.text(-1,31,'u/double_dose_larry', size=12)
plt.tight_layout()
plt.savefig('war_pos_heatmap.png')
```
