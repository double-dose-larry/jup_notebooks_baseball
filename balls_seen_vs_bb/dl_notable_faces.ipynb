```python
import pybaseball as pb
from bs4 import BeautifulSoup
from urllib.request import urlretrieve, urlopen
import os
```

```python
NOTABLES = """Juan Soto
Shohei Ohtani
Joey Gallo
Brandon Nimmo
Josh Donaldson
Corey Seager
Cavan Biggio
Jake Fraley
Carter Keiboom
Tim Anderson
Valdimir Guererro Jr.
Mitch Haniger
J.P. Crawford
Yasmani Grandal
Bryce Harper
Robbie Grossman
Fernando Tatis Jr.
Marcus Semien
Willians Astudillo
Salvador Perez
Zack Collins
Kevin Pillar
Isiah Kiner-Falefa
Whit Merrifield
Joey Votto
Aaron Judge""".split("\n")
```

```python
for player in NOTABLES:
    print("getting", player, "face")
    first, last  = player.split(" ")[:2]
    bref_id = pb.playerid_lookup(last, first, fuzzy=True)["key_bbref"].iloc[0]
    print(bref_id)
    bref_url = f"https://www.baseball-reference.com/players/{bref_id[0]}/{bref_id}.shtml"
    soup = BeautifulSoup(urlopen(bref_url).read())
    img = soup.select('#meta img')[0]["src"]
    file_path = f"notable_faces/{player}.jpg"
    if not os.path.exists(file_path):
        urlretrieve(img, file_path)
    print("done")
```
